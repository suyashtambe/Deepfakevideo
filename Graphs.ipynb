```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"video_features_statistical.csv")
```

```python
# Set up the plotting environment
sns.set(style="whitegrid")
plt.figure(figsize=(15, 12))

# 1. Distribution Plot for Each Feature
plt.subplot(3, 2, 1)
sns.histplot(df, x='mean_intensity', hue='label', kde=True, element="step")
plt.title('Mean Intensity Distribution')

plt.subplot(3, 2, 2)
sns.histplot(df, x='optical_flow_mean', hue='label', kde=True, element="step")
plt.title('Optical Flow Mean Distribution')

plt.subplot(3, 2, 3)
sns.histplot(df, x='optical_flow_variance', hue='label', kde=True, element="step")
plt.title('Optical Flow Variance Distribution')

# 2. Box Plots to Compare Real and Fake for Each Feature
plt.subplot(3, 2, 4)
sns.boxplot(data=df, x='label', y='mean_intensity')
plt.title('Mean Intensity by Label')

plt.subplot(3, 2, 5)
sns.boxplot(data=df, x='label', y='optical_flow_mean')
plt.title('Optical Flow Mean by Label')

plt.subplot(3, 2, 6)
sns.boxplot(data=df, x='label', y='optical_flow_variance')
plt.title('Optical Flow Variance by Label')

plt.tight_layout()
plt.show()

# 3. Pair Plot to Show Relationships between Features
sns.pairplot(df, hue='label', diag_kind='kde')
plt.suptitle('Pairwise Relationships between Features', y=1.02)
plt.show()

# 4. Correlation Heatmap for All Features
plt.figure(figsize=(8, 6))
correlation_matrix = df[['mean_intensity', 'optical_flow_mean', 'optical_flow_variance']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```
